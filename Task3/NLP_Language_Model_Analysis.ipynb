```python
from transformers import pipeline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load BERT sentiment analysis model
classifier = pipeline("sentiment-analysis")

print("Language Model Loaded Successfully")
```

```python
texts = [
    "I love artificial intelligence and machine learning.",
    "This internship is very challenging but interesting.",
    "I hate when my code doesn't run properly.",
    "Machine learning is transforming the future.",
    "This task is extremely boring."
]

results = classifier(texts)

for text, result in zip(texts, results):
    print("Text:", text)
    print("Prediction:", result)
    print()
```

```python
test_sentences = [
    "This project is amazing!",
    "I hate debugging code.",
    "The meeting is scheduled tomorrow.",
    "AI will change the world.",
    "I am very disappointed with the results."
]

analysis_results = classifier(test_sentences)

for sentence, result in zip(test_sentences, analysis_results):
    print("Sentence:", sentence)
    print("Sentiment:", result['label'])
    print("Confidence Score:", result['score'])
    print()
```

```python
data = pd.DataFrame(analysis_results)
data["Sentence"] = test_sentences

data
```

```python
plt.figure(figsize=(8,5))

plt.bar(data["label"], data["score"])

plt.title("Sentiment Prediction Confidence")
plt.xlabel("Sentiment")
plt.ylabel("Confidence Score")

plt.show()
```

```python
## Observations

#The BERT language model performs well when identifying clear
#positive and negative sentiments in sentences. It assigns a
#confidence score indicating how certain the model is about
#its prediction.

#However, the model may struggle with neutral or ambiguous
#sentences where sentiment is not clearly expressed.
```
